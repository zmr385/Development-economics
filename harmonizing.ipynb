```python
# We want to harmonize the files 
import pyreadstat
import os
import pandas as pd
import numpy as np
```

```python
# now we want to read in the files 
folder = "/Users/albertolsen/Documents/Stud.Polit./11. Bologna/Development Economics/Data/"

#creating placeholder variable
label_list = []

for files in os.listdir(folder):
    if files.endswith(".sav") and any(sub in files for sub in ["r3", "r4", "r5", "r6", "r7"]):      # sub here does not mean anything. Just a 'variable'
        sav_path = os.path.join(folder, files)
        df, meta = pyreadstat.read_sav(sav_path)
        if any(sub in files for sub in["r3"]): 
            r3 = meta
        if any(sub in files for sub in["r4"]): 
            r4 = meta
        if any(sub in files for sub in["r5"]): 
            r5 = meta
        if any(sub in files for sub in["r6"]): 
            r6 = meta
        if any(sub in files for sub in["r7"]):
            r7 = meta


# I create 7 different lists with the column names and column labels
r7_meta = r7.column_labels
#r7_meta.append(r7.column_names) 

r6_meta = r6.column_labels
#r6_meta.append(r6.column_names)

r5_meta = r5.column_labels
#r5_meta.append(r5.column_names)

r4_meta = r4.column_labels
#r4_meta.append(r4.column_names)

r3_meta = r3.column_labels
#r3_meta.append(r3.column_names)

# we add an element to the beginning of the list to identify the round. 
r3_meta.insert(0, "Round 3")        # where to put it and then what to put. (where to put, what to put)
r4_meta.insert(0, "Round 4")
r5_meta.insert(0, "Round 5")
r6_meta.insert(0, "Round 6")
r7_meta.insert(0, "Round 7")    


# for easier working all the labels i put them in a list
all_rounds = [r3_meta, r4_meta, r5_meta, r6_meta, r7_meta]
```

```python
# Now we want to change everything into upper case
for round in all_rounds:        # this works because I call an object that already exists outside the loop. 
    for i in range(len(round)):             # we define the range as the length of the list. Doable if you state from 0. Otherwise need range(x,y)
        round[i] = round[i].upper()         # here we use square brackets to tackle all the elements in the list. 

print(all_rounds)
```

```python
# now I want to remove the question numbers. 
# the basic code is: 
# variable.split(". ", 1)[1]  # so I split at the first encountered ". " inside () and then I choose the second part. [1].

only_labels = all_rounds.copy()

# I want to do it in a loop. 
only_labels = all_rounds.copy()

for rounds in only_labels: 
    for i in range(len(rounds)):
        rounds[i] = rounds[i].split(". ", 1)[1] if ". " in rounds[i] else rounds[i]     # here I use a condition to check if there is a ". " in the label. If there is, I split it and take the second part. If there isn't, I keep the label as it is.

print(only_labels[0])


```

```python
# we store the labels for each round seperately

round_3 = only_labels[0]
round_4 = only_labels[1]
round_5 = only_labels[2]
round_6 = only_labels[3]
round_7 = only_labels[4]

# Finding common elements in all five rounds
common_labels = set(round_3).intersection(round_4, round_5, round_6, round_7)
print(common_labels)
print(len(common_labels))
```

```python
# I create an indicator function that I use to store the final values to
indicator = []

for y in [round_3,round_4, round_5, round_6, round_7]: 
    sr = [1 if x in common_labels else 0 for x in y ]
    indicator.append(sr)

print(indicator)
# so this should, I hope, have created a list that we can use to filter the dataframes.
```

```python
# we want to know whether people trust the president. And for some reason the questions has been left out by mistake. We add it manually. 
# Round 3: Q55a
# Round 4: Q49a
# Round 5: Q59a
# Round 6: Q52a
# Round 7: Q43a

# Here you can use the loc function and add a 1. 
pres_trust = [next((i for i, label in enumerate(r3_meta) if "PRESIDENT" and "TRUST" in label), None)]
pres_trust.append(next((i for i, label in enumerate(r4_meta) if "PRESIDENT" and "TRUST" in label), None))
pres_trust.append(next((i for i, label in enumerate(r5_meta) if "PRESIDENT" and "TRUST" in label), None))
pres_trust.append(next((i for i, label in enumerate(r6_meta) if "PRESIDENT" and "TRUST" in label), None))
pres_trust.append(next((i for i, label in enumerate(r7_meta) if "PRESIDENT" and "TRUST" in label), None))

print(pres_trust)

indicator[0][pres_trust[0]]= 1
indicator[1][pres_trust[1]]= 1
indicator[2][pres_trust[2]]= 1
indicator[3][pres_trust[3]]= 1
indicator[4][pres_trust[4]]= 1

```

```python
# We store the modified indicator function independently. Now the place where the "trust the president" is added. 
i_3 = indicator[0][1:]  # Extract the first row and remove the first column
i_4 = indicator[1][1:]
i_5 = indicator[2][1:]
i_6 = indicator[3][1:]
i_7 = indicator[4][1:]

print(i_3)
print(i_3[pres_trust[0]])
print(sum(i_3))
```

```python
# We write a function to remove the unwanted columns 
def remove_columns(df): 
    while 0 in df[1]:                   # while = så længe 0 eksisterer i 2. række af df gøres det her
        del df[0][df[1].index(0)]       #   vi sletter fra række 1, den værdi som matcher indekset for det første 0 i række 1. 
        del df[1][df[1].index(0)]
    
    # print(len(df[0]))     de her print viser blot at det virker
    # print(df[1])
    if 0 not in df[1]:
       print(len(df[0]))
 
    return(df[0])        
```
