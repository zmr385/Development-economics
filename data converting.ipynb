```python
# So we want to change all the sav files into csv files in our directory
folder = "/Users/albertolsen/Documents/Stud.Polit./11. Bologna/Development Economics/Data/"

for files in os.listdir(folder):
    if files.endswith(".sav"):          # only looking at sav files
    
        sav_path = os.path.join(folder, files)          # creating path for all the files
        df, meta = pyreadstat.read_sav(sav_path)        # loading the sav file 
        csv_path = os.path.join(folder,files.replace(".sav", ".csv"))   # creating the path for names for csv files from the sav file path
        df.to_csv(csv_path, index=False)            # index = false because we don't want a useless column at the beginning of the data file

# Here we export to CSV files. It is difficult to store meta data, so we would like to export to .dta files too. 

```
